### METRIC CALCULATION

```python
def score_cal(ground_truth, gpt_output):
    true_positives = 0.0
    false_positives = 0.0
    false_negatives = 0.0
    true_positives += len(set(ground_truth) & set(gpt_output))
    false_positives += len(set(gpt_output) - set(ground_truth))
    false_negatives += len(set(ground_truth) - set(gpt_output))
    if (true_positives + false_positives) != 0: precision = true_positives / (true_positives + false_positives)
    else : precision = 0
    if (true_positives + false_negatives) != 0: recall = true_positives / (true_positives + false_negatives)
    else : recall = 0
    if (precision + recall) != 0: f_score = 2 * (precision * recall) / (precision + recall)
    else : f_score = 0
    return [precision, recall, f_score]
```
